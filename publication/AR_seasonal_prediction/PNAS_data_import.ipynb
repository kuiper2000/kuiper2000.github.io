## **PNAS2020**


This script is an example file showing how to import atmospheric river data used in the paper : ***Seasonal Prediction of western North America Atmospheric Rivers from Github***. One should notice that due to the storage limitation of Github and policy of GFDL, all files are saved with a post-process form (i.e., AR). If you are interested in the raw data AR (e.g., u, v and q), please contact corresponding author: kaichiht@princeton.edu
